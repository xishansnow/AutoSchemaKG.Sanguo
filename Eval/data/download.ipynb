```python
import os
from datasets import load_dataset

# 1. Định nghĩa đường dẫn thư mục mong muốn
# Sử dụng raw string (r'...') để tránh lỗi ký tự đặc biệt trong đường dẫn Windows
output_dir = r"C:\Users\heheh\Desktop\UET\GenAI\AutoSchemaKG\Eval\data"

# 2. Tạo thư mục nếu chưa tồn tại (quan trọng để tránh lỗi FileNotFoundError)
os.makedirs(output_dir, exist_ok=True)

# 3. Tải dataset
print("Đang tải dataset MedQuAD...")
ds = load_dataset("lavita/MedQuAD")

# 4. Định nghĩa tên file và đường dẫn đầy đủ
file_name = "medquad.csv"
full_path = os.path.join(output_dir, file_name)

# 5. Lưu phần 'train' ra file CSV
# (MedQuAD thường chứa dữ liệu chính trong key 'train')
if 'train' in ds:
    ds['train'].to_csv(full_path)
    print(f"Đã lưu thành công tại: {full_path}")
else:
    # Trường hợp dataset không chia split, lưu trực tiếp
    ds.to_csv(full_path)
    print(f"Đã lưu thành công tại: {full_path}")
```
